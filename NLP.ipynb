```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'textID': [1, 2, 3],
    'text': ['hello', 'good', 'bad'],
    'sentiment': ['positive', 'neutral', 'negative']
})

# Correct column selection=
df = df[['text', 'sentiment']]
print(df)
```

```python
df.columns
```

```python
df.head()
```

```python
df=df.dropna(subset=['text'])
```

```python
df.shape
```

```python
df['text'] = df['text'].astype(str)
```

```python
type('text')
```

```python
df['text_lower'] =df['text'].str.lower()
```

```python
print(df[['text','text_lower']])
```

```python
df['text_upper']=df['text'].str.upper()
```

```python
print(df[['text','text_upper']])
```

```python
df.to_csv("processed_amazon.csv",index=False)
```
